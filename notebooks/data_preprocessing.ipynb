```python
dropped_na_df = data2.dropna(subset=["bid_price_1","bid_price_2" ,"bid_price_3","bid_price_4", "bid_price_5"])
```

```python
dropped_na_df.head()
```

```python
# Create the new column with the specified condition
dropped_na_df['is_bid_strictly_decreasing'] = (
    (dropped_na_df['bid_price_1'] >= dropped_na_df['bid_price_2']) &
    (dropped_na_df['bid_price_2'] >= dropped_na_df['bid_price_3']) &
    (dropped_na_df['bid_price_3'] >= dropped_na_df['bid_price_4']) &
    (dropped_na_df['bid_price_4'] >= dropped_na_df['bid_price_5'])
)
```

```python
# Create the new column with the specified condition
dropped_na_df['is_ask_strictly_increasing'] = (
    (dropped_na_df['ask_price_1'] <= dropped_na_df['ask_price_2']) &
    (dropped_na_df['ask_price_2'] <= dropped_na_df['ask_price_3']) &
    (dropped_na_df['ask_price_3'] <= dropped_na_df['ask_price_4']) &
    (dropped_na_df['ask_price_4'] <= dropped_na_df['ask_price_5'])
)
```
